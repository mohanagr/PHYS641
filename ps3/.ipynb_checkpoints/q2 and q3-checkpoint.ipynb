```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
import tools
```

```python
hdul=fits.open("advact_tt_patch.fits")
mymap=hdul[0].data
hdul.close()
mymap=np.asarray(mymap,dtype='float')
nx,ny = mymap.shape
width=1024 # chopping off a big enough part of the map due to save RAM. 
#This is enough to account for large scale noise we care about for detecting stuff close to Abel cluster
```

```python
plt.imshow(mymap[nx//2-width:nx//2+width,ny//2-width:ny//2+width],aspect='equal')
plt.colorbar()
patch=mymap[nx//2-width:nx//2+width,ny//2-width:ny//2+width].copy()
```

```python
smooth_ps = tools.estimate_ps(patch,sig_smooth=100)
```

```python
plt.imshow(np.log10(np.fft.fftshift(smooth_ps[:,:])),aspect='auto',vmax=12)
plt.colorbar()
```
