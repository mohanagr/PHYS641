```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
import tools
```

```python
hdul=fits.open("advact_tt_patch.fits")
mymap=hdul[0].data
hdul.close()
mymap=np.asarray(mymap,dtype='float')
nx,ny = mymap.shape
width=512 # chopping off a big enough part of the map to save RAM. 
```

```python
plt.imshow(mymap[nx//2-width:nx//2+width,ny//2-width:ny//2+width],aspect='equal')
plt.colorbar()
patch=mymap[nx//2-width:nx//2+width,ny//2-width:ny//2+width].copy()
```

```python
fitp=[-269.05607134,    0.38579781,   -1.32372561,    6.47035581,   -5.68035932] #use best fit params
templ1=tools.gauss2d(fitp,512)
plt.imshow(templ1,aspect='equal')
```

Subtract of a temporary model for good PS estimate.

```python
res=patch-templ1
smooth_ps = tools.estimate_ps(res,sig_smooth=10)
```

```python
patchft = np.fft.fft2(tools.pad_map(patch))
templ=tools.gauss2d(fitp,1024)
tft = np.fft.fft2(templ)
Ninvt = tft/smooth_ps

matched_filt_rhs = np.abs(Ninvt.shape[0]**2*np.real(np.fft.ifft2(np.conj(Ninvt)*patchft)))
 # normalization factors of N^2 added, as explained in Q4
```

```python
lhs = np.sum(templ*(Ninvt.shape[0]**2)*np.real(np.fft.ifft2(Ninvt)))  # normalization factors of N^2 added, as explained in Q4
```

```python
mf = matched_filt_rhs[:patch.shape[0],:patch.shape[1]]/lhs
```

```python
errbar = np.sqrt(1/lhs)
print("Expected error bar is:", errbar)
```

## FIRST LET'S MAKE SURE WE SEE ABEL CLUSTER IN THE DEAD CENTER

```python
width=64
nx,ny=mf.shape
plt.imshow(mf[nx//2-width:nx//2+width,ny//2-width:ny//2+width],aspect='equal')
plt.colorbar()
```

```python
mini_patch=mf[nx//2-width:nx//2+width,ny//2-width:ny//2+width].copy()
expected_noise = np.std(mini_patch[:40,:])
```

```python
print("Expected noise is:", expected_noise)
```

```python
mf=mf/expected_noise #get SNR to look for other galaxy clusters
```

```python
width=64
nx,ny=mf.shape
plt.imshow(mf[nx//2-width:nx//2+width,ny//2-width:ny//2+width],aspect='equal')
plt.colorbar()
plt.title("SNR OF ABEL CLUSTER")
```
