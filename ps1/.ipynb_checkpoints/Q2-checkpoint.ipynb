```python
k=np.arange(1,1000)
```

We see in the following plots that as $k$ (or $n$ in Jon's question) increases, the approximation becomes more and more exact.

```python
s=3
lmbda=(np.sqrt(s**2+4*k)-s)**2/4
prob_poisson=st.poisson.pmf(k,lmbda)
prob_gauss=np.exp(-s**2/2)/np.sqrt(2*np.pi*lmbda)
plt.plot(k,prob_poisson/prob_gauss)
plt.title("Ratio of Poisson/Gaussian for 3sigma")
plt.xlabel("k")
plt.xlim(0,100)
z=prob_poisson/prob_gauss
print("We need k at least", k[np.argmin(np.abs(z-2))])
```

```python
s=5
lmbda=(np.sqrt(s**2+4*k)-s)**2/4
prob_poisson=st.poisson.pmf(k,lmbda)
prob_gauss=np.exp(-s**2/2)/np.sqrt(2*np.pi*lmbda)
plt.plot(k,prob_poisson/prob_gauss)
plt.title("Ratio of Poisson/Gaussian for 5sigma")
plt.xlabel("k")
plt.xlim(0,1000)
z=prob_poisson/prob_gauss
print("We need k at least", k[np.argmin(np.abs(z-2))])
```
