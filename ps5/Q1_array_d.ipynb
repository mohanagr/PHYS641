```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def get_uvw(arr='d',decl=34.0784*np.pi/180,freq=1.4e9):
    
    # decl: source decl at which uvw coordinates will be centered
    # by default it's the zenith of the array
    # Returns: baseline in wavelengths in Earth's center frame, projection matrix to convert it to uvw frame
    
    if arr=='a':
        array=np.loadtxt("./vla_a_array.txt")
    else:
        array=np.loadtxt("./vla_d_array.txt")
    xyz=array[:,:3]*3e8*1e-9
    N=xyz.shape[0]

    #take phi=0 at the longitude of the array. little bit arbitrary, could as well have fixed East as Greenwich.
    #Will have to subract (or add?) the longitude of the array from phi if you wanted the correct hour angle.

    east=[0,1,0]
    zen = [np.cos(decl),0,np.sin(decl)]
    north=np.cross(zen,east)

    proj_mat = np.vstack([east,north,zen]).T

    nb = N*(N-1)//2
    
    print("Num baselines = ", nb)

    uvw=np.zeros((nb,3))

    count=0
    for i in range(N):
        for j in range(i+1,N):
            uvw[count,:] = xyz[i,:]-xyz[j,:] #Baselines in m, in Earth Center ref frame. will project later
            count+=1
    print(f"Max baseline for array {arr} is {np.max(np.sqrt(np.sum(uvw**2,axis=1))):4.2f} m")
    uvw=np.vstack([uvw,-uvw])*freq/3e8 # since image is real, it's FT needs to bbe hermitian conjugate F*(-k) = F(k) = 1
    return xyz,uvw,proj_mat
```

```python
xyz,uvw,proj_mat=get_uvw(arr='d')
xyz_local = xyz@proj_mat
uvw_local = uvw@proj_mat
print(f"Vertical scatter in array d is {np.std(xyz_local[:2]):4.2f} m")
f=plt.gcf()
f.set_size_inches(10,4)
plt.subplot(121)
plt.scatter(xyz_local[:,0],xyz_local[:,1])
plt.xlabel("EW")
plt.ylabel("NS")

plt.subplot(122)
plt.scatter(uvw_local[:,0],uvw_local[:,1])
plt.xlabel("u")
plt.ylabel("v")

plt.tight_layout()
```

```python
def grid_uv(uv,res,npix):
    
    #
    fov = res*npix
    delu = 1/fov
    
    # umax = 1/res since our resolution is smaller than what's needed for better plotting, we'll never reach npix
    # so no risk of exceeding array indices. Max uv will be around 830 as shown below
    
    S = np.zeros((2*npix,2*npix),dtype='float64')
    
#     print(delu)
#     uv_gridded=np.round(uv/delu).astype(int)
    uv_int=np.asarray(uv/delu,dtype='int')
    for i in range(uv.shape[0]):
        S[uv_int[i,0],uv_int[i,1]] = S[uv_int[i,0],uv_int[i,1]] + 1   #rows are U, columns are V
    
    return S
```

```python
# Array a
lm = 3e8/1.4e9 #m
d = 25 #m dish size
res = lm/1032 #max baseline
fov = lm/d
print("Array A")
print(f"Synth. resol\t: {np.rad2deg(res)*3600:4.2f}\"\nFOV\t\t: {np.rad2deg(fov)*3600:4.2f}\"\nMax UV\t\t: {int(fov/res):d}")

#let's keep pixel resolution more than FWHM of synthesized image so we can see the FWHM clearly in the plot.
#this basically means max_npix >= max_u, so that we don't saturate the grid.

npix=2048
myres=fov/npix
print(f"Pixel resol \t: {np.rad2deg(myres)*3600:4.2f}\"")
```

```python
uv_grid = grid_uv(uvw_local[:,:2],myres,npix)
```

```python
img = np.abs(np.fft.ifft2(uv_grid))**2
```

```python
f=plt.gcf()
f.set_size_inches(5,5)
w=100
x0=img.shape[0]//2
pixres = np.rad2deg(myres)*3600
myext = [-w*pixres,w*pixres,w*pixres,-w*pixres]
plt.imshow(np.fft.fftshift(img)[x0-w:x0+w,x0-w:x0+w],extent=myext,interpolation='none')
plt.xlabel("Declination (arcsec)")
plt.ylabel("RA (arcsec)")
# plt.xlim()
# plt.ylim(x0-25,x0+25)
```

```python
f=plt.gcf()
f.set_size_inches(10,4)
xx=np.fft.fftshift(img)[x0-50:x0+50,x0].copy()
xx=xx/xx.max()
plt.subplot(121)
plt.plot(xx)
plt.title("RA")
mygauss = lambda x,amp,mu,sig: amp*np.exp(-0.5*(x-mu)**2/sig**2)
fitp,cov=curve_fit(mygauss,np.arange(0,100),xx,[1,50,10])
print(f"FWHM along RA {np.rad2deg(fitp[2]*2.355*myres)*3600:4.2f} arcsec")

xx=np.fft.fftshift(img)[x0,x0-50:x0+50].copy()
xx=xx/xx.max()
plt.subplot(122)
plt.plot(xx)
plt.title("Declination")
mygauss = lambda x,amp,mu,sig: amp*np.exp(-0.5*(x-mu)**2/sig**2)
fitp,cov=curve_fit(mygauss,np.arange(0,100),xx,[1,50,10])
print(f"FWHM along dec {np.rad2deg(fitp[2]*2.355*myres)*3600:4.2f} arcsec")
```

### VLA website states that in case of natural weighting (all visiblities having the same weight, as is the case here), FWHM must be multiplied by 1.5.  https://science.nrao.edu/facilities/vla/docs/manuals/oss/performance/resolution

#### Stated FWHM = 46''. Therefore, for natural weighting we expect FWHM ~ 70''. Thus, approximately, the my FWHM matches with what's stated on the website.

### Rotation synthesis

Zenith

```python
t=np.linspace(-4,4,51)
phi = t*np.pi/12
#2*pi in 24 hrs.
xyz,uvw,proj_mat=get_uvw(arr='d')

nn=uvw.shape[0]
uv_big = np.zeros((nn*t.shape[0],3))
for i,p in enumerate(phi):
    rot_mat = np.asarray([[np.cos(p),-np.sin(p),0],[np.sin(p),np.cos(p),0],[0,0,1]]).T 
    #our vectors are rows, so taking a Transpose. Result is same as what Jon uses.
    uv_big[i*nn:(i+1)*nn,:] = uvw@rot_mat
print(uv_big.shape)
uv_big[:] = uv_big@proj_mat
```

```python
uv_grid = grid_uv(uv_big[:,:2],myres,npix)
img = np.abs(np.fft.ifft2(uv_grid))**2
f=plt.gcf()
f.set_size_inches(10,5)

plt.subplot(121)
plt.scatter(uv_big[:,0],uv_big[:,1])
plt.title("A array, 8 hrs UV coverage, zenith")
plt.xlabel('v')
plt.ylabel('u')

plt.subplot(122)
w=100
x0=img.shape[0]//2
plt.imshow(np.fft.fftshift(img)[x0-w:x0+w,x0-w:x0+w],interpolation='none')
plt.title("Source at zenith")

plt.tight_layout()
# plt.xlim()
# plt.ylim(x0-25,x0+25)
```
